2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
df.columns
```

```python
yes_customers = df[df["response"] == "Yes"]
```

```python
analysis = (
    yes_customers
    .groupby(["policy_type", "gender"])
    .agg({
        "monthly_premium_auto": "mean",
        "customer_lifetime_value": "mean",
        "total_claim_amount": "mean"
    })
    .round(2)
)

print(analysis)
```

Overall, Corporate Auto (Male) appears to be the most profitable and lowest-risk segment because it combines relatively high customer lifetime value with the lowest claim costs. Special Auto (Male) is also an attractive segment due to its high lifetime value and relatively low claims. In contrast, Personal Auto customers—especially males—tend to have higher claim amounts, making them comparatively riskier despite generating higher premiums in some cases.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
state_customers = (
    df.groupby("state")
      .size()
      .reset_index(name="customer_count")

)

states_over_500 = state_customers[state_customers["customer_count"] > 500]
states_over_500
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_analysis = (
    df.groupby(["education", "gender"])
      .agg(
          max_clv=("customer_lifetime_value", "max"),
          min_clv=("customer_lifetime_value", "min"),
          median_clv=("customer_lifetime_value", "median")
      )
      .round(2)
)

clv_analysis
```

Overall, customers with a High School or Below education appear to be the most valuable segment based on both the highest median and highest maximum customer lifetime values. In contrast, customers with a Doctor degree have the lowest customer lifetime values overall, making them the least valuable segment in this analysis.
